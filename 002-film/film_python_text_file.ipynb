```python
class Film:
    movie_name = "default"
    movie_type = "default"
    movie_date = "default"
    def __init__(self, n,t,d) -> None:
        self.movie_name = n
        self.movie_type = t
        self.movie_date = d
        pass
    def __str__(self) -> str:
        return f"Film ismi:{self.movie_name}|Film turu:{self.movie_type}|Film tarihi:{self.movie_date}"
        pass
    
    
class FilmArsivi:
    file_name = "film_arsivi.txt"
    
    def __init__(self) -> None:
        pass
    def addMovie(self, movie):
        with open(self.file_name, "a") as f:
            f.write(f"{movie.__str__()}\n")
        pass
    def findMovie(self, movie_name) -> bool:
        with open(self.file_name, "r") as f:
            text = f.read()
            lines = text.split("\n")
            try:
                for line in lines:
                    if movie_name == line.split("|")[0].split(":")[1]:
                        return True
            except IndexError:
                pass
        return False
    
    def __str__(self) -> str:
        with open(self.file_name, "r") as f:
            text = f.read()
        return text

film_arsivi = FilmArsivi()

while True:
    try:
        secim = input("1-Film Ekle\n2-Film Ara\n3-Cikis\n")
        secim = int(secim)
    except ValueError:
        print(f"Gecersiz giris: {secim}")
        continue
        
    if secim == 1:
        fi = input("Film ismini giriniz:")
        ft = input("Film turunu giriniz:")
        fd = input("Film tarihini giriniz:")
        try:
            yeni_film = Film(fi, ft, fd)
            film_arsivi.addMovie(yeni_film)
            print("++Film Eklendi++")
            continue
        except Exception as e:
            print(f"--HATA: {e}--\n")
            continue

    elif secim == 2:
        fi = input("Film ismini giriniz:")
        if film_arsivi.findMovie(fi):
            print("Film listenizde bulunmaktadir.")
            continue
        else:
            print("Film bulunamadi. Ismi kontrol edip tekrar deneyiniz.")
            continue

    elif secim == 3:
        print("Cikis yapiliyor")
        break
    else:
        print(f"Gecersiz giris: {secim}")
        continue

     
        
    
    
    
```
